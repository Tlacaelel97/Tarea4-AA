# Imports

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

# Compara el rendimiento de clasificadores de SVM con kernel coseno, lineal y RBF usando 3 repeticiones de validación cruzada estratificada de 5 particiones.

## Carga del dataset

```python
newsgroups_train = fetch_20newsgroups(subset='train', 
                                      remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test', 
                                     remove=('headers', 'footers', 'quotes'))
```

## Configuración de TF-IDF con Preprocesamiento

Con `stop_words='english'` filtramos palabras vacías comunes, `norm='l2'` normaliza los histogramas resultantes y `min_df=5` ignora palabras que aparecen en menos de 5 documentos.


```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', norm='l2', min_df=5)

# Representación de las bolsas de palabras
X_train_tfidf = tfidf_vectorizer.fit_transform(newsgroups_train.data)
X_test_tfidf = tfidf_vectorizer.transform(newsgroups_test.data)

y_train = newsgroups_train.target
y_test = newsgroups_test.target

print(f"Dimensiones de la matriz TF-IDF (Entrenamiento): {X_train_tfidf.shape}")
print(f"Número de clases: {len(newsgroups_train.target_names)}")
```
